```python
import math
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize

plt.rcParams['figure.facecolor'] = 'lightgrey'
plt.rcParams['figure.figsize'] = (12, 6)
plt.rcParams['axes.grid'] = True
plt.rcParams['grid.alpha'] = 0.5
plt.rcParams['grid.linestyle'] = '--'
plt.rcParams['axes.titlesize'] = 16
plt.rcParams['axes.labelsize'] = 12
plt.rcParams['legend.frameon'] = True
plt.rcParams['legend.facecolor'] = 'white'
plt.rcParams['legend.edgecolor'] = 'black'

colors = ["cornflowerblue", "indianred", "darkseagreen", "plum", "dimgray"]
```

```python
def fx(K, lam, So):
     return (lam ** K * So ** (K - 1) * np.exp(-lam * So)) / math.factorial(K - 1)

def PBL(spread):
     p = 0.5 - 0.08 * spread
     return np.maximum(0, p)
```

```python
K = 60
lam = 3
So = np.linspace(10, 33, 1000)
```

```python
plt.figure()
plt.plot(So, fx(K, lam, So), color=colors[0])
plt.title(f'Price Distribution')
plt.xlabel('So')
plt.ylabel('PDF')
plt.show()
```

```python
spread = np.linspace(0, 6, 100)

fun = 0.6 * PBL(spread) * spread
```

```python
plt.figure()
plt.plot(spread, spread, color=colors[0], label='Linear Payoff')
plt.plot(spread, 0.6 * spread, color=colors[1], label='60% Payoff')
plt.plot(spread, fun, color=colors[2], label='Custom Payoff')
plt.title('Payoff Functions')
plt.xlabel('Spread')
plt.ylabel('Payoff')
plt.legend()
plt.show()
```
