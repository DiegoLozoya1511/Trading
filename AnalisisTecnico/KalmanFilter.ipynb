```python
import numpy as np
import matplotlib.pyplot as plt
import visualization
from typing import Tuple
```

```python
def generate_linear_data() -> Tuple[np.ndarray, np.ndarray]:
    '''
    Generate random linear data to test our Kalman Filter implemmentation
    '''
    x = np.linspace(0, 10, 100)
    y = 1.8 * x -3.21 + np.random.normal(0, 1, 100)
    return x, y

def generate_quad_data() -> Tuple[np.ndarray, np.ndarray]:
    '''
    Generate random quadratic data to test our Kalman Filter implemmentation
    '''
    x = np.linspace(0, 10, 100)
    y = 1.8 *x - 0.7 * x**2 -3.21 + np.random.normal(0, 1, 100)
    return x, y
```

```python
xl, yl = generate_linear_data()
plt.scatter(xl, yl)
plt.title("Random Linear Data")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```
